```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data from our report card generator:
data = {
    'Name': ['lucky', 'lokesh', 'lalit', 'kunal', 'lakshay'],
    'Math': [85, 72, 91, 68, 77],
    'Science': [90, 78, 88, 72, 84],
    'English': [82, 85, 79, 70, 88],
    'Total': [257, 235, 258, 210, 249],
    'Average': [85.7, 78.3, 86.0, 70.0, 83.0],
    'Grade': ['A', 'B', 'A', 'C', 'B']
}

df = pd.DataFrame(data)

sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))
plt.figure(figsize=(8, 6))
grade_counts = df['Grade'].value_counts()
colors = sns.color_palette('pastel')[0:len(grade_counts)]
plt.pie(grade_counts, 
        labels=grade_counts.index, 
        colors=colors,
        autopct='%.0f%%',
        startangle=90,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1})
plt.title('Grade Distribution (Percentage)', fontsize=14, pad=20)
plt.tight_layout()
plt.savefig('grade_piechart.png')
plt.show()
print("""
Pie Chart Observation:
- 40% of students received A grades
- 40% received B grades
- Only 20% (1 student) received a C grade
- The majority (80%) of students achieved above-average grades (A/B)
""")
```

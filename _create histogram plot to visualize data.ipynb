```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data from our report card generator
data = {
    'Name': ['lucky', 'lokesh', 'lalit', 'kunal', 'lakshay'],
    'Math': [85, 72, 91, 68, 77],
    'Science': [90, 78, 88, 72, 84],
    'English': [82, 85, 79, 70, 88],
    'Total': [257, 235, 258, 210, 249],
    'Average': [85.7, 78.3, 86.0, 70.0, 83.0],
    'Grade': ['A', 'B', 'C', 'B', 'A']
}

df = pd.DataFrame(data)

sns.set_style("whitegrid")
plt.figure(figsize=(10, 8))
plt.figure(figsize=(10, 8))
sns.histplot(df['Average'], bins=5, kde=True, color='red')
plt.title('Distribution of Average Scores', fontsize=14, pad=20)
plt.xlabel('Average Score (0-100)', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.axvline(df['Average'].mean(), color='green', linestyle='--', label=f'Mean: {df["Average"].mean():.1f}')
plt.legend()
plt.tight_layout()
plt.savefig('average_distribution.png')
plt.show()
print("""
Histogram Observation:
The average scores cluster between 70-90, with most students scoring above 75 and also score 60.
The distribution is slightly left-skewed with one student scoring exactly 70.
The mean average is 80.0, indicating generally good class performance.
""")
```

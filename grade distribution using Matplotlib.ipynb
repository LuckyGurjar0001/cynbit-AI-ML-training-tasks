```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data ={
    "Name":["lucky","lokesh","lalit","kunal","lakshay"],
    "Grade":["A","B","A","C","B"]
}
df = pd.DataFrame(data)
```

```python
grade_counts = df['Grade'].value_counts().sort_index()
```

```python
plt.figure(figsize=(8,5))
plt.bar(grade_counts.index,grade_counts.values,colour='skyblue')
```

```python
plt.title("Grade distribution(Matplotlib)",fontsize=15)
plt.xlabel("Grade",fontsize = 13)
plt.ylabel("Number of students",fontsize=13)
plt.grid(axis='y',linestyle='--',alpha=0.7)
for i,count in enumerate(grade_counts.value):
    plt.text(i,count + 0.2,str(count),ha='center', fontsize=13)
plt.show()
             
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your DataFrame)
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Grade": ["A", "B", "A", "C", "B"]
}
df = pd.DataFrame(data)

# Count grades
grade_counts = df['Grade'].value_counts().sort_index()

# Create bar chart
plt.figure(figsize=(8, 5))
plt.bar(grade_counts.index, grade_counts.values, color='skyblue')

# Customize
plt.title("Grade Distribution (Matplotlib)", fontsize=14)
plt.xlabel("Grade", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add counts on bars
for i, count in enumerate(grade_counts.values):
    plt.text(i, count + 0.2, str(count), ha='center', fontsize=12)

plt.show()
```

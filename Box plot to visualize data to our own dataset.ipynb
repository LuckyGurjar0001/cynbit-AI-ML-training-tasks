```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data from our report card generator
data = {
    'Name': ['lucky', 'lokesh', 'lalit', 'kunal', 'lakshay'],
    'Math': [85, 72, 91, 68, 77],
    'Science': [90, 78, 88, 72, 84],
    'English': [82, 85, 79, 70, 88],
    'Total': [257, 235, 258, 210, 249],
    'Average': [85.7, 78.3, 86.0, 70.0, 83.0],
    'Grade': ['A', 'B', 'A', 'A', 'B']
}

df = pd.DataFrame(data)

sns.set_style("whitegrid")
plt.figure(figsize=(10, 10))
plt.figure(figsize=(10, 10))
sns.boxplot(data=df[['Math', 'Science', 'English']], palette="Set2")
plt.title('Subject-wise Score Distribution', fontsize=14, pad=20)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Marks', fontsize=12)
plt.tight_layout()
plt.savefig('subject_performance.png')
plt.show()
print("""
Box Plot Observation:
avaerage score 75-90.
Science has the highest median score (84) and least variability.
Math shows the widest range of scores (70-91), indicating varying student abilities.
English has the most consistent performance with most scores between 79-88.
""")
```

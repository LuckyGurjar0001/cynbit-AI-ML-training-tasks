```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data from our report card generator
data = {
    'Name': ['lucky', 'lokesh', 'lalit', 'kunal', 'lakshay'],
    'Math': [85, 72, 91, 68, 77],
    'Science': [90, 78, 88, 72, 84],
    'English': [82, 85, 79, 70, 88],
    'Total': [257, 235, 258, 210, 249],
    'Average': [85.7, 78.3, 86.0, 70.0, 83.0],
    'Grade': ['A', 'B', 'A', 'C', 'B']
}

df = pd.DataFrame(data)

# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
plt.figure(figsize=(12, 6))
for index, row in df.iterrows():
    plt.plot(['Math', 'Science', 'English'], row[['Math', 'Science', 'English']], 
             marker='o', label=row['Name'])
plt.title('Individual Student Performance Across Subjects', fontsize=14, pad=20)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Marks', fontsize=12)
plt.legend()
plt.tight_layout()
plt.savefig('individual_performance.png')
plt.show()
print("""
Line Plot Observation:
lalit shows strong performance in Math/Science but relatively weaker in English.
kunal is consistently the lowest performer across all subjects.
lakshay shows remarkable improvement from Math to English.
""")
```
